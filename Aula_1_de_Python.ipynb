```python
df.columns = [
    'ano',
    'senioridade',
    'contrato',
    'cargo',
    'salario',
    'moeda',
    'usd',
    'residencia',
    'remoto',
    'empresa',
    'tamanho_empresa'
]
print("Colunas traduzidas:")
display(df.columns)
```

```python
print("Categorias originais da coluna 'senioridade':")
df["senioridade"].value_counts()
```

```python
df["contrato"].value_counts()
```

```python
df["remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade)

print("Categorias traduzidas da coluna 'senioridade':")
display(df['senioridade'].value_counts())
```

```python
contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(contrato)

print("Categorias traduzidas da coluna 'contrato':")
display(df['contrato'].value_counts())
```

```python
tamanho_empresa = {
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)

print("Categorias traduzidas da coluna 'tamanho_empresa':")
display(df['tamanho_empresa'].value_counts())
```

```python
remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(remoto)

print("Categorias traduzidas da coluna 'remoto':")
display(df['remoto'].value_counts())
```
